Adicionando bibliotecas

```python
import requests
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
from lxml import etree, html
import matplotlib.pyplot as plt
```

#Importando os links externos necessários

```python
start_url = 'https://pokemondb.net/'
url = start_url+'pokedex/all'
```

# Fazendo a solicitação dos dados em formato de texto


```python
headers = { 
    'User-Agent'      : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/71.0.3578.98 Safari/537.36', 
    'Accept'          : 'text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8', 
    'Accept-Language' : 'en-US,en;q=0.5',
    'DNT'             : '1', # Do Not Track Request Header 
    'Connection'      : 'close'
}
```

```python
data = requests.get(url, headers=headers, timeout=5).text
soup = BeautifulSoup(data,"html.parser")
```

# Encotrando o pokemon pela tag table


```python
table = soup.find('table') # achando do nome do pokemon pela table pokedex
```

#definido o data frame


```python
df = pd.DataFrame(columns=['Numero',  'Nome', 'Tipo', 'Total', 'HP', 'Ataque', 'Defesa',  'Ataque Especial',  'Defesa Especial', 'Velocidade'])
```

# Pegando os dados do table principal

```python
# Obtendo todas as linhas da tabela
for row in table.tbody.find_all('tr'): # em html uma linha da tabela é representada pela tag <tr>
    # Obtendo todas as colunas em cada linha
    columns = row.find_all('td')  # em html uma coluna da tabela é representada pela tag <td>
    if(columns != []):
        numero = columns[0].text.strip(' ')
        nome = columns[1].text.strip(' ')
        tipo = columns[2].text.strip(' ')
        total = columns[3].text.strip(' ')
        hp = columns[4].text.strip(' ')
        ataque = columns[5].text.strip(' ')
        defesa = columns[6].text.strip(' ')
        especial_ataque = columns[7].text.strip(' ')
        especial_defesa = columns[8].text.strip(' ')
        velocidade = columns[9].text.strip(' ')

        print('Numero - id: ' + numero)
        print('Nome: ' + nome)
        print('Tipo: ' + tipo)
        print('Total: ' + tipo)
        print('HP: ' + hp)
        print('Ataque: ' + ataque)
        print('Defesa: ' + defesa)
        print('Ataque Especial: ' + especial_ataque)
        print('Defesa Especial: ' + especial_defesa)
        print('Velocidade: ' + velocidade)
        print('-------------------------------')

        df = pd.concat([df, pd.DataFrame.from_records([{'Numero': numero,  'Nome': nome, 'Tipo': tipo, 'Total': total, 'HP': hp, 'Ataque': ataque, 'Defesa': defesa, 'Ataque Especial': especial_ataque, 'Defesa Especial': especial_defesa, 'Velocidade': velocidade,}])], ignore_index=True)
```
